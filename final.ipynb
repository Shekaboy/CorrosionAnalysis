```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("EXP.csv")
df = df.drop(["Year","First Author","Ref" ,"Alloy, condition/form" ,"Composition, wt%"],axis=1)
df.head()
```

```python
print("Unique values in 'c' column{}\nUnique values in 'M' column {} \n does 'CR,\\nmm/y' have null values? {}\n Does 'Exp' have null values{}\nDoes 'TA,\\n%' have null values{}".format( df["c"].unique() , df["M"].unique() , np.unique(df["CR,\nmm/y"].isna(),return_counts = True),np.unique(df["Exp"].isna(),return_counts=True),np.unique(df["TA,\n%"].isna(),return_counts=True)) )
```

```python
df_ = df[df['c'].notna()]
df_ = df_[df_['TA,\n%'].notna()]
df_ = df_[df_['Exp'].notna()]
df_ = df_[df_['M'].notna()]
df_ = df_[df_['CR,\nmm/y'].notna()]
print("Removed {} rows due to null values".format(len(df) - len(df_)))
```

```python
def _(x:str):
    x = x.replace(".",",")
    x = x.replace("l","1")
    x = x.replace("c1","d")
    x = x.replace("I","1")
    x = x.replace("O","0")
    x = x.split(",")[-1]
    x = x.replace(" ","")
    x = x.replace("ud","-1")  
    x = x.replace("nd","-1")
    x = x.replace("d","*24*60*60")
    x = x.replace("h","*60*60")
    x = x.replace("m","*60")
    x = x.replace("w","*7*24*60*60")
    x = x.replace("y","365**24*60*60")
    x = x.replace("s","")
    x = x.replace("?","-1")
    return int(str(eval(x)).replace("j",""))

df_["Exp"] = df_["Exp"].apply(_)
df_ = df_[df_.Exp != -1]
```
